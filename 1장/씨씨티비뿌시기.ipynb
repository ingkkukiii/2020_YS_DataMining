### <font color ='red'> 문자열자료형 -> 숫자자료형으로 바꾸는 작업 필요

```python
list(PS2.columns)
```

```python
#for c in list(PS2.columns):
for c in list(PS2.columns)[1:]: #0번 구별 제외, c는 4번 반복
    PS2[c] = PS2[c].apply(lambda x : x.replace(',','')) #대괄호 안에 문자 -> 열 숫자 -> 행
    PS2[c] = pd.to_numeric(PS2[c]) #자료형타입으로 변환 
```
